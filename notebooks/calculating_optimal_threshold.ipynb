```python
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, f1_score

from notebook_utils import best_f1_threshold, get_validation_df
from src.utils import models_dir
```

```python
threshold, f1 = best_f1_threshold("Namibia_North_2020")
print(f"Threshold: {threshold}")
print(f"F1: {f1}")
```

```python
df = get_validation_df("Namibia_North_September_2020")
```

```python
thresholds = [0.5, 0.01]
fig, axes = plt.subplots(1, len(thresholds), figsize=(15,5))
y_true = df["y_true"]
for i, threshold in enumerate(thresholds):
    y_pred = df["y_pred_decimal"] > threshold
    
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non-crop", "Crop"])
    disp.plot(ax=axes[i])
    axes[i].set_title(
        f"Threshold: {threshold}\n" + 
        f"F1: {round(f1_score(y_true, y_pred), 3)}\n" +
        f"Accuracy: {round(accuracy_score(y_true, y_pred), 3)}"
    )


plt.tight_layout()
```
